Introduction

In this project, the objective was to generate synthetic data for 1,000 individuals using the Faker library. The generation of artificial data is an important step in data analysis and software development, as it allows simulation of real-world scenarios without using sensitive or personal information.

An object-oriented approach was applied by creating a Person class, which defines key attributes such as name, age, address, phone number, and job. The generated objects were then transformed into a pandas DataFrame, enabling efficient data handling, analysis, and further processing.

```python
# 1. Install
!pip install faker pandas
```

```python
# 2. Import
from faker import Faker
import pandas as pd
import random

fake = Faker()
```

```python
# 3. Make the Class
class Person:
    def __init__(self):
        self.name = fake.name()
        self.age = random.randint(18, 70)
        self.address = fake.address()
        self.phone = fake.phone_number()
        self.job = fake.job()
    def to_dict(self):
        return {
              "Name": self.name,
              "Age": self.age,
              "Address": self.address,
              "Phone": self.phone,
              "Job": self.job
          }
```

```python
# 4. Generate
people = [Person().to_dict() for _ in range(1000)]
```

```python
# 5. Save
df = pd.DataFrame(people)
```

```python
# 6. Check
df.head()
```

```python
df.to_csv("fake_people.csv", index=False)
```
